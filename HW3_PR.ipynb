Linear system: Let us consider a linear model $y = Wx + b$, where $W$ is an $m \times d$ matrix and $b \in \mathbb{R}$. Write an expression for the optimal values of $W$ and $b$ that minimize the following cost function:

$J(W, b) = \frac{1}{2} \|y - Wx - b\|^2 $

To find the optimal values, we need to set the gradients of the cost function with respect to the optimization variables to zero. Let's start by deriving the expressions for the gradients:

1. Gradient with respect to $W$:

$\frac{\partial J}{\partial W_{ij}} = \frac{\partial}{\partial W_{ij}} \left(\frac{1}{2} \|y - Wx - b\|^2\right)$


Using the chain rule, we have:

$\frac{\partial J}{\partial W_{ij}} = \frac{\partial J}{\partial \left(y - Wx - b\right)} \cdot \frac{\partial \left(y - Wx - b\right)}{\partial W_{ij}}$


Since $\frac{\partial \left(y - Wx - b\right)}{\partial W_{ij}} = -x_{ij}$, we can simplify the expression:

$\frac{\partial J}{\partial W_{ij}} = -(y - Wx - b) \cdot x_{ij}$


2. Gradient with respect to $b$:

$\frac{\partial J}{\partial b} = \frac{\partial}{\partial b} \left(\frac{1}{2} \|y - Wx - b\|^2\right)$


Using the chain rule again, we have:

$\frac{\partial J}{\partial b} = \frac{\partial J}{\partial \left(y - Wx - b\right)} \cdot \frac{\partial \left(y - Wx - b\right)}{\partial b}$


Since $\frac{\partial \left(y - Wx - b\right)}{\partial b} = -1$, we can simplify the expression:

$\frac{\partial J}{\partial b} = -(y - Wx - b)$

Now, to find the optimal solution, we set the gradients to zero:

1. Set $\frac{\partial J}{\partial W_{ij}} = 0$:

$-(y - Wx - b) \cdot x_{ij} = 0$


2. Set $\frac{\partial J}{\partial b} = 0$:

$-(y - Wx - b) = 0$


Simplifying the equations above, we get the following expressions for the optimal values:

1. Optimal $W$:

$W = (X^T X)^{-1} X^T y$


where $X$ is the matrix of input samples with each row corresponding to a sample, and $y$ is the vector of target values.

2. Optimal $b$:

$b = \bar{y} - \bar{W} \cdot \bar{x}$


where $\bar{y}$ is the mean of the target values, $\bar{W}$ is the mean of the optimal $W$ values, and $\bar{x}$ is the mean of the input samples.

Let us consider the following linear model $y = Wx + b$, where $W$ is a Toeplitz matrix corresponding to a filter $w$ of length $k$. We can write the $d + k - 1 \times d$ Toeplitz matrix as:


W =
\begin{bmatrix}
w_1 & 0 & 0 & \dots & 0 & 0 \\
w_2 & w_1 & 0 & \dots & 0 & 0 \\
w_3 & w_2 & w_1 & \dots & 0 & 0 \\
\vdots & \vdots & \vdots & \ddots & \vdots & \vdots \\
w_k & w_{k-1} & w_{k-2} & \dots & \dots & w_1 \\
0 & w_k & w_{k-1} & \dots & \dots & 0 \\
0 & 0 & w_k & \dots & \dots & 0 \\
\vdots & \vdots & \vdots & \ddots & \vdots & \vdots \\
0 & 0 & 0 & \dots & w_k & w_{k-1} \\
0 & 0 & 0 & \dots & 0 & w_k \\
\end{bmatrix}


We can also define the linear model above as $y = w * x + b$, where $x * w$ denotes the convolution of two vectors that can be defined as $(x * w)_j = \sum_{i=1}^d x_i w_{j-i+1}$, where $w_i$ denotes the $i$th entry in vector $w$, and $w_i = 0$ for $i < 1$ and $i > k$. With our definition above, we assume that $j$ is only valid for $1 \leq j \leq d + k - 1$.

The cost function is given by:
$$J(W, b) = \frac{1}{2} \|y - w * x - b\|^2 \tag{1}$$

Now, let's derive the expressions for $\frac{\partial J}{\partial w_i}$ and $\frac{\partial J}{\partial b}$:

1. Gradient with respect to $w_i$:

$$\frac{\partial J}{\partial w_i} = \frac{\partial}{\partial w_i} \left(\frac{1}{2} \|y - w * x - b\|^2\right)$$


Using the chain rule, we have:

$$\frac{\partial J}{\partial w_i} = \frac{\partial J}{\partial (y - w * x - b)} \cdot \frac{\partial (y - w * x - b)}{\partial w_i}$$


Since $\frac{\partial (y - w * x - b)}{\partial w_i} = -x_i \delta(i - j + 1)$, where $\delta$ is the Kronecker delta, we can simplify the expression:

$$\frac{\partial J}{\partial w_i} = -(y - w * x - b) \cdot x_i \delta(i - j + 1)$$


2. Gradient with respect to $b$:

$$\frac{\partial J}{\partial b} = \frac{\partial}{\partial b} \left(\frac{1}{2} \|y - w * x - b\|^2\right)$$


Using the chain rule, we have:

$$\frac{\partial J}{\partial b} = \frac{\partial J}{\partial (y - w * x - b)} \cdot \frac{\partial (y - w * x - b)}{\partial b}$$


Since $\frac{\partial (y - w * x - b)}{\partial b} = -1$, we can simplify the expression:

$$\frac{\partial J}{\partial b} = -(y - w * x - b)$$


To find the optimal values, we set the gradients to zero:

1. Set $\frac{\partial J}{\partial w_i} = 0$:

$$-(y - w * x - b) \cdot x_i \delta(i - j + 1) = 0$$


2. Set $\frac{\partial J}{\partial b} = 0$:

$$-(y - w * x - b) = 0$$


Simplifying the equations above, we get the following expressions for the optimal values:

1. Optimal $w$:

$$w = (X^T X)^{-1} X^T y$$


where $X$ is the matrix of input samples with each row corresponding to a sample, and $y$ is the vector of target values.

2. Optimal $b$:

$$b = \bar{y} - \bar{w} * \bar{x}$$


where $\bar{y}$ is the mean of the target values, $\bar{w}$ is the mean of the optimal $w$ values, and $\bar{x}$ is the mean of the input samples.

(a) To correctly classify the XOR data using a 2-layer neural network with ReLU nonlinearity, we can choose the following values for weights and biases:

W1 = [[1, -1], [-1, 1]]
b1 = [-0.5, -0.5]

(b) The decision boundary for the first node in layer 1, which corresponds to w1_11x(1) + w1_12x(2) + b1_1 = 0, can be plotted as a line with the equation x(2) = x(1) + 0.5.

(c) The decision boundary for the second node in layer 1, which corresponds to w1_21x(1) + w1_22x(2) + b1_2 = 0, can be plotted as a line with the equation x(2) = x(1) - 0.5.

(d) The decision boundary for the output node in layer 2, which corresponds to w2_11y1_1 + w2_12y1_2 + b2_1 = 0, can be plotted as a line with the equation y1_2 = -y1_1.



```python
import numpy as np
import matplotlib.pyplot as plt

# XOR data
x = np.array([[-1, -1], [-1, 1], [1, -1], [1, 1]])
y = np.array([1, 0, 0, 1])

# Decision boundary for the first node in layer 1
x_decision = np.linspace(-2, 2, 100)
y_decision = x_decision + 0.5

plt.scatter(x[:, 0], x[:, 1], c=y)
plt.plot(x_decision, y_decision, color='red', label='Decision boundary for Node 1 in Layer 1')
plt.xlabel('x(1)')
plt.ylabel('x(2)')
plt.legend()
plt.title('XOR Data with Decision Boundaries')
plt.show()
```

```python
# Decision boundary for the second node in layer 1
x_decision = np.linspace(-2, 2, 100)
y_decision = x_decision - 0.5

plt.scatter(x[:, 0], x[:, 1], c=y)
plt.plot(x_decision, y_decision, color='green', label='Decision boundary for Node 2 in Layer 1')
plt.xlabel('x(1)')
plt.ylabel('x(2)')
plt.legend()
plt.title('XOR Data with Decision Boundaries')
plt.show()
```

```python
# Decision boundary for the output node in layer 2
x_decision = np.linspace(-2, 2, 100)
y_decision = -x_decision

plt.scatter(x[:, 0], x[:, 1], c=y)
plt.plot(x_decision, y_decision, color='blue', label='Decision boundary for Output Node in Layer 2')
plt.xlabel('y1_1')
plt.ylabel('y1_2')
plt.legend()
plt.title('XOR Data with Decision Boundaries')
plt.show()
```
